```python
import os 
from nltk.util import ngrams
def get_ngram(n):
  directory =os.getcwd() +'/'
  filelist = os.listdir(directory)
#print(filelist)
  for file in filelist:
    with open(os.getcwd()+'/'+file,'r')as f:
      sentence = f.read()
    #print(sentence)

      characters = [char for char in sentence]
    #print(characters)
      ng = ngrams(characters,n)
      x=[''.join (ngramed) for ngramed in ng]
      print(x)
```

```python
get_ngram(3)
```

```python
# jc = C n Y / X U Y
def jaccard(x,y):
  numerator = x.intersection(y)
  denomenator = x.union(y)
  return len(numerator)/len(denomenator)

doc3 = open(os.getcwd()+'/doc3.txt','r')
sentence = doc3.read()
print(sentence)

characters = [char for char in sentence]

ngramed = ngrams(characters , 3)
x=[''.join (n_g) for n_g in ngramed]
print(x)

doc4 = open(os.getcwd()+'/doc4.txt','r')
sentence2 = doc4.read()
print(sentence2)

characters2 = [char for char in sentence2]

ngramed2 = ngrams(characters2 , 3)
y=[''.join (n_g1) for n_g1 in ngramed2]
print(y)


jaccard(set(x),set(y))
```

```python

def soundex(name):
 name = name.upper()
 soundex = ""
 soundex += name[0]
 dictionary = { "BFPV" :"1" , "CGJKQSXZ":"2" , "DT":"3" , "L":"4" , "MN":"5"  , "R":"6", "AEIOUHWY":"0" }

 for char in name[1:] :
     for key in dictionary.keys():
          if char in key:
            code = dictionary[key]
            #print(code)
            if code != soundex[-1]:
              soundex+=code
              #print(soundex)
              soundex = soundex.replace("0","")
              soundex = soundex[:4].ljust(4,"0")

 return(soundex)


```
